## UpSet

Based on the idea "UpSetR" (https://github.com/hms-dbmi/UpSetR)

Two versions are available : PyUpset and PyUpsetHue

### PyUpset

PyUpset(df, key, value, unique=True, intersection=True, addvalue=True,
        bcolor="black", griddots=False, lsize=12, maxwidth=10, row_clustering=False,
        row_dendogram=False, method='average', metric='euclidean', 
        boxratio=(None, None), spacers=(.1, .1), gridkwargs={})
        
Draw an upset graph based on a dataframe for which **key** is the usual index column and **value** is the column name for categories (see example below).

**unique** : Either you want to draw intersections including only one category or not.

**intersection** : If you want to make clear intersection (Ture) or overlapped values between intersections (False).

**addvalue** : Add count value on the top of the bars

**bcolor** : Bar color

**griddots** : Either a grid dotes (True) or an heatmap (False). Heatmap is recommanded if you increase the maxwidth size

**lsize** : Label size

**maxwidth** : Maximum number of intersections to draw

**row_clustering** : Whether to reorganize categorized using clustering

**row_dendogram** : Draw dendogram based on clustering results

**method**, **metric** : Method and metric to use for clustering, look at [scipy linkage method](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.cluster.hierarchy.linkage.html) for infos.

**boxratio**, **spacers**, **gridkwargs** : GridSpec arguments for the plot.

```python
import seahorse as sh

from seahorse.core import constants
constants.SHOWMODE = True

import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/hms-dbmi/UpSetR/master/inst/extdata/movies.csv", sep=";", header=0)

# Clean up the data

df = df[[column for column in df.columns if column not in ["Watches", "ReleaseDate"]]]
df = pd.melt(df, id_vars=["Name", "AvgRating"], var_name="Kind", value_name="Found")
df = df[df["Found"] == 1].drop("Found", axis=1)
df["AvgRating"] = df["AvgRating"] // 1

df.head()
```

```python
graph = sh.PyUpset(df, key="Name", value="Kind", griddots=True)
```

```python
graph = sh.PyUpset(df, key="Name", value="Kind", griddots=False, maxwidth=30, bcolor="Grey")
```

```python
graph = sh.PyUpset(df, key="Name", value="Kind", griddots=True, maxwidth=10, unique=False, addvalue=False)
```

```python
graph = sh.PyUpset(df, key="Name", value="Kind", griddots=True, maxwidth=10, unique=False, addvalue=True, row_dendogram=True)
```

PyUpsetHue(df, key, value, hue, nvalue=0, unique=True, intersection=True,
           griddots=False, lsize=12, maxwidth=10, palette=None, addvalue=True,
           boxratio=(None, None), spacers=(.1, .1), gridkwargs={}, legend=True)
           
Split the intersection values based on a hue column name.

**palette** : Color palette to use, by default the seaborn default color palette is used

Other arguments are similar to PyUpset ones.

```python
def categorise_rates(rate) :
	if rate < 3 : return "low"
	if rate > 3 : return "high"
	return "medium"

df["AvgRating"] = df["AvgRating"].apply(categorise_rates)

graph = sh.PyUpsetHue(df, key="Name", value="Kind", hue="AvgRating", griddots=True)
```
