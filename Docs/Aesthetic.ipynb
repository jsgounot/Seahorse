```python
from seahorse import Graph, SContainer
from seahorse import gwrap

import numpy as np
import pandas as pd
```

### Labels, title and ticks

```python
df = pd.DataFrame(np.random.randn(1000, 4) * 100000, index=pd.date_range('1/1/2000', periods=1000), 
                  columns=list('ABCD'))
df = df.cumsum()

graph = Graph(df)
graph.df.plot()
```

```python
graph = Graph(df)
graph.df.plot()
graph.ax.set_title("Graph title", size=18)
graph.set_labels("Xlabel", "YLabel", size=16)
```

```python
graph = Graph(df)
graph.df.plot()
graph.thousand_sep(axis="y", separator=",")
```

```python
df = gwrap.sns.load_dataset("tips")
graph = Graph(df)
graph.sns.barplot("sex", "tip")
graph.transform_xticks(new_names={"Female" : "Girls"}, style="italic", size=20)
```
